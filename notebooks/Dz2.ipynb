```python
def count_unique(str):
    keys = []
    for i in str:
        if not(i in keys):
            keys.append(i)
            
    slovar = dict.fromkeys(keys)
    
    for key in slovar.keys():
        slovar[key] = 0
    
    for i in str:
        slovar[i] += 1
        
    return slovar

str = input()
slovar = count_unique(str)
print(slovar)
```

```python
def read_csv(file):
    mas = []
    with open(file, "r") as f:
        for line in f:
            mas.append(line.split(","))
    return mas[0], mas[1:]
file = input()
read_csv(file)
```
